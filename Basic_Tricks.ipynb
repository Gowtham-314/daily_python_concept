Use _ (underscore) for “I don’t care” values

```python
for _ in range(5):
    print("Hello, World!")
```

One-Line If–Else (Ternary Operator)

```python
marks = 45
result = "Pass" if marks >= 35 else "Fail"
print(result)
```

Count Elements Easily (Hidden Power)

```python
nums = [1, 2, 2, 3, 2]
print(nums.count(2))
```

Reverse Anything in One Line

```python
text = "JavaScript"
print(text[::-1])  # Output: tpircSavaJ
```

Default Dictionary Value (Very Useful Trick)

```python
student = {"name": "Gowtham"}
print(student.get("age",0))
```

Multiple Assignment (Very Pythonic)

```python
a, b, c = 5, 10, 15
print(a, b, c)  # Output: 15 5 10
```
