```*args and **kwargs```

````python
def func(*a):
    print(type(a))
    return a

func(5,10)
````

````python
def func(*a):
    return sum(a)

func(5,10)
````

````python
func(5,10,15)
````

````python
func(*[5,10,15,20])
# How ?
# * is used to unpack the list and pass each element as a separate argument to the function. So, func(*[5,10,15,20]) is equivalent to func(5,10,15,20). 
````

````python
temp=[]


def student_info(**kwargs):
    temp.append(kwargs)

    return temp
````

````python
student_info(name="John", age=25, grade="A")
````

````python
student_info(name="John", age=35, grade="B")
````
