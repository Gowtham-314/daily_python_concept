Variable Length Agruments

Using *args 

```python
def tot_sum(*numbers):
    result=0
    for num in numbers:
        result+=num
    return result

print(tot_sum(1,2,3,4,5,6))
```

Using **Kwargs

```python
def stu_info(**details):
    for key,value in details.items():
        print(f"{key} : {value}")
        
stu_info(name='Gokul',age=20)
```

lambda Function

```python
x=int(input("Enter the Number:"))
squr=lambda x:x**2
print(squr(x))
```
