```python
import shelve
with shelve.open('mydata.db') as db:
    db['username'] = 'admin'
    db['password'] = 'secret'
print('Data stored successfully.')
```

```python
import shelve
with shelve.open('mydata.db') as db:
    print('Username:', db['username'])
    print('Password:', db['password'])
```

```python
with shelve.open('mydata.db') as db:
    db["score"]=[100,200,300]
print("Score list stored successfully.")
```

```python
with shelve.open('mydata.db') as db:
    for key in db:
        print(f'{key}: {db[key]}')
```
